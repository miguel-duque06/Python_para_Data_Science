# DATA VIZUALIZATION


```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/imigrantes_canada.csv')
```

```python
dados
```

```python
dados.info()
```

```python
dados.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980,2014)))
```

```python
anos
```

```python
brasil = dados.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

## Conhecendo a biblioteca Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])  # Plota linha ano vs imigrantes
plt.title('Imigração para o Canadá')  # Define título do gráfico
plt.xlabel('Ano')  # Rótulo eixo X
plt.ylabel('Número de Imigrantes')  # Rótulo eixo Y (corrigido)
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])  # Ticks específicos X
plt.show()  # Exibe gráfico
```

## Criando figuras com Matplotlib


```python
fig, ax = plt.subplots(figsize=(8,4))  # Cria figura + eixos com tamanho 8x4
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])  # Plota linha no eixo ax
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  # Ticks a cada 5 unidades no eixo X
ax.set_title('Imigração do Brasil para o Canadá\n1980 - 2013')  # Título com quebra de linha
ax.set_xlabel('Ano')  # Rótulo eixo X
ax.set_ylabel('Número de imigrantes')  # Rótulo eixo Y
plt.show()
```

* Criando subplots

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))  # Cria figura com 1 linha x 2 colunas de subplots

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 - 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()  # Adiciona grade ao gráfico

axs[1].boxplot(dados_brasil['imigrantes'])  # Cria boxplot dos dados
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 - 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,8))  # Cria 2x2 subplots com tamanho 12x8
fig.subplots_adjust(hspace=0.8, wspace=0.6)  # Ajusta espaçamento entre subplots
fig.suptitle('Imigração dos quatro maiores países da América dos sul para o Canadá\n1980 - 2013')  # Título geral da figura

axs[0,0].plot(dados.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(dados.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(dados.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(dados.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)  # Define limite Y uniforme para todos gráficos

plt.show()
```

## Costumizando com Matplotlib

* Alterando as cores

```python
america_sul = dados.query('Região == "América do Sul"')
```

```python
america_sul = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []
for pais in america_sul.index:  # Itera sobre países da América do Sul
  if pais == 'Brasil':
    cores.append('green')  # Brasil em verde
  else:
    cores.append('silver')  # Outros em cinza

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)  # barh= barra horizontal
ax.set_title('América do sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá de 1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

* Adicionando anotações

```python
cores = []
for pais in america_sul.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('América do sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá de 1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul['Total']):  # Adiciona valores nas barras
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')  # ha= alinhamento horizontal, va= alinhamento vertical

ax.set_frame_on(False)  # Remove borda da figura
ax.get_xaxis().set_visible(False)  # Esconde eixo X
ax.tick_params(axis='both', which='both', length=0)  # Remove ticks das duas direções

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data = top_10, x='Total', y=top_10.index, orient='h')

ax.set_title('Países com maior imigração para o Canadá\n1980 - 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

* Alterando as cores com paletas

```python
def gerar_grafico_paleta(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h',
              palette=palette)
  ax.set_title('Países com maior imigração para o Canadá\n1980 - 2013', loc='left', fontsize=18)
  ax.set_xlabel('Número de imigrantes', fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('coolwarm')
```

```python
gerar_grafico_paleta('tab10')
```

* Explorando temas

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes')
fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle': -45})

fig.show()
```

* Personalizando vizualizações

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração do Brasil para o Canadá\n1980 - 2013')
fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle': -45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')

fig.show()
```

* Alterando cores

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração do Brasil para o Canadá\n1980 - 2013')
fig.update_traces(line_color='green', line_width=4)
fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle': -45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')

fig.show()
```
